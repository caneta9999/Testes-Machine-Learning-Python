#Carregando dados

```python
from sklearn.preprocessing import StandardScaler
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
from sklearn.datasets import make_blobs
```

```python
data, _ = make_blobs(n_samples=2500, centers=5, cluster_std=3)
data
```

```python
grafico = px.scatter(x = data[:,0], y = data[:,1])
grafico.show()
```

#Kmeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
```

```python
grafico1 = px.scatter(x = data[:,0], y = data[:,1], color=kmeans.labels_)
grafico2 = px.scatter(x = kmeans.cluster_centers_[:,0], y = kmeans.cluster_centers_[:,1], size = [25, 25, 25])
grafico_geral = go.Figure(data = grafico1.data + grafico2.data)
grafico_geral.show()
```

# MeanShift

```python
from sklearn.cluster import MeanShift
```

```python
meanShift = MeanShift()
meanShift.fit(data)
```

```python
labels = len(pd.unique(meanShift.labels_))
labels
```

```python
meanShift.cluster_centers_
```

```python
grafico1 = px.scatter(x = data[:,0], y = data[:,1], color=kmeans.labels_)
grafico2 = px.scatter(x = meanShift.cluster_centers_[:,0], y = meanShift.cluster_centers_[:,1], size = [25]*pd.unique(meanShift.labels_))
grafico_geral = go.Figure(data = grafico1.data + grafico2.data)
grafico_geral.show()
```
