```python
import pandas as pd
```

```python
dados = pd.read_csv('Employee.csv')
dados.head(50)
```

```python
X_dados = dados.iloc[:,0:8]
X_dados
```

```python
y_dados = dados.iloc[:,8:]
y_dados
```

```python
from sklearn import preprocessing
labelEncoder = preprocessing.LabelEncoder()
```

```python
from sklearn.preprocessing import OneHotEncoder
oneHotEncoder = OneHotEncoder()
```

```python
X_dados['Education'] = labelEncoder.fit_transform(X_dados['Education'])
X_dados['City'] = labelEncoder.fit_transform(X_dados['City'])
X_dados['Gender'] = labelEncoder.fit_transform(X_dados['Gender'])
X_dados['EverBenched'] = labelEncoder.fit_transform(X_dados['EverBenched'])
```

```python
X_dados
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
oneHotEncoder = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [0,2,5,6])], remainder='passthrough')
```

```python
X_dados = oneHotEncoder.fit_transform(X_dados)
```

```python
X_dados
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_dados = scaler.fit_transform(X_dados)
X_dados
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
modelo = keras.Sequential([ layers.Dense(58, activation="relu"), layers.Dense(54, activation="relu"), layers.Dense(50, activation="relu")])
```

```python
modelo.compile(optimizer='Adamax', metrics='accuracy', loss='sparse_categorical_crossentropy')
```

```python
modelo.fit(X_dados, y_dados, epochs=50)
```
